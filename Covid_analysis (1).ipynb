```python
#Importing libraries
import pandas as pd
import sqlite3
```

```python
#Importing SQL Engine
from sqlalchemy import create_engine
engine = create_engine('sqlite://' ,echo = False)
```

```python
#Reading dataset 
covid = pd.read_excel('C:/Users/Sam/Downloads/covid-data.xlsx')
vaccine= pd.read_excel('C:/Users/Sam/Downloads/vaccinedata.xlsx')
covid.head(50)

```

```python
print(vaccine.columns.to_list())
```

```python
#Connecting data with sql engine
covid.to_sql('coviddata', con = engine)
vaccine.to_sql('vaccine1data', con = engine)
```

```python
# Having a glance at the dataset
sql =  '''

select * from coviddata
where continent is not null

'''
df_sql = pd.read_sql_query(sql, con = engine)
df_sql.head(10)
```

```python
# total deaths vs total cases (Percentage)
sql = '''

Select Location, date, total_cases,total_deaths,round((total_deaths/total_cases)*100,4) as DeathPercentage
From coviddata
Where location like 'India'
and continent is not null 
order by 1,2

'''
df_sql = pd.read_sql_query(sql, con = engine)
df_sql.tail(50)
```

```python
# Calculating total percentage of population infected.
sql = '''

Select Location, date, Population, total_cases,  (total_cases/population)*100 as PercentPopulationInfected
From coviddata
Where location like 'India'
order by 1,2

'''
df_sql = pd.read_sql_query(sql, con = engine)
df_sql.tail(30)
```

```python
# Calculating highest percentage of population infected
sql = '''

Select Location, Population, MAX(total_cases) as HighestInfectionCount,  Max((total_cases/population))*100 as PercentPopulationInfected
From coviddata
--Where location like '%states%'
Group by Location, Population
order by PercentPopulationInfected desc

'''
df_sql = pd.read_sql_query(sql, con = engine)
df_sql.head(30)
```

```python
#Total death count
sql = '''

Select Location, MAX(cast(Total_deaths as int)) as TotalDeathCount
From coviddata
--Where location like '%states%'
Where continent is not null 
Group by Location
order by TotalDeathCount desc

'''
df_sql = pd.read_sql_query(sql, con = engine)
df_sql.head(30)
```

```python
#Total deaths per continent
sql = '''

Select continent, MAX(cast(Total_deaths as int)) as TotalDeathCount
From coviddata
Where continent is not null 
Group by continent
order by TotalDeathCount desc

'''
df_sql = pd.read_sql_query(sql, con = engine)
df_sql.head(30)
```

```python
#Total death percentage
sql= '''

Select SUM(new_cases) as total_cases, SUM(cast(new_deaths as int)) as total_deaths, SUM(cast(new_deaths as int))/SUM(New_Cases)*100 as DeathPercentage
From coviddata
where continent is not null 
order by 1,2

'''

df_sql = pd.read_sql_query(sql, con = engine)
df_sql.head(30)
```

```python
# Shows Percentage of Population that has recieved at least one Covid Vaccine Using CTE.
sql = '''

With PopvsVac (Continent, Location, Date, Population, New_Vaccinations, RollingPeopleVaccinated)
as
(
Select dea.continent, dea.location, dea.date, dea.population, vac.new_vaccinations
, SUM(vac.new_vaccinations) OVER (Partition by dea.Location Order by dea.location, dea.Date) as RollingPeopleVaccinated
--, (RollingPeopleVaccinated/population)*100
From coviddata dea
Join vaccine1data vac
	On dea.location = vac.location
	and dea.date = vac.date
where dea.continent is not null 
--order by 2,3
)
Select *, (RollingPeopleVaccinated/Population)*100
From PopvsVac
'''
df_sql = pd.read_sql_query(sql, con = engine)
df_sql.head(1000)
```

```python
#Death population by continent.
sql = '''

select continent,population, max(total_cases) as totalcases, max(cast(total_deaths as int)) as totaldeath, round(MAX((total_deaths/population)*100),2) as deathpop
from coviddata
where continent is not null
group by 1
order by 2 desc

'''
df_sql = pd.read_sql_query(sql, con = engine)
df_sql.head(30)
```

```python
#total vaccinations across continents
sql = '''
Select cd.continent,cd.population, sum(cv.new_vaccinations) as totalvaccinations from coviddata cd
join vaccine1data cv on
cd.location = cv.location
where cd.continent is not null
group by 1


'''
df_sql = pd.read_sql_query(sql, con = engine)
df_sql.head(15)

```

```python
# displays first value for locations in the dataset.
sql = '''
select location, total_cases,new_cases, first_value(new_cases) over (partition by location order by total_cases desc) as firstvalue
from coviddata 
where continent is not null 

'''
df_sql = pd.read_sql_query(sql, con = engine)
df_sql.head(1000)
```
